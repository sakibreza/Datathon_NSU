# Pre-Processing of data
- data from .txt to chucks of .csv 
- concating of .csv chunks 
- droping the missing values
- final preprocessed data in .csv

```python
import pandas as pd

folder = 'raw_data'

data1= pd.read_csv(folder+'/train1.csv')

train_data1= pd.DataFrame(data1)
# print(train_data1.info())

##
data2= pd.read_csv(folder+'/train_df2.csv',header=None)

train_data2= pd.DataFrame(data2)
# print(train_data2.info())

data3= pd.read_csv(folder+'/train_df333.csv',header=None)

train_data3= pd.DataFrame(data3)
# print(train_data3.info())

data4= pd.read_csv(folder+'/train_df44.csv',header=None)

train_data4= pd.DataFrame(data4)
# print(train_data4.info())
final = pd.concat([train_data1, train_data2 , train_data3 , train_data4])
# print(final.info())
```

```python
train_data1.head()
data1 = train_data1.iloc[:,:-5]
data1.columns = ['id','activity','timestamp','x','y','z']
data1.head()
```

```python
train_data2.head()
data2 = train_data2
data2.columns = ['id','activity','timestamp','x','y','z']
data2.head()
```

```python
train_data3.head()
data3 = train_data3
data3.columns = ['id','activity','timestamp','x','y','z']
data3.head()
```

```python
train_data4.head()
data4 = train_data4
data4.columns = ['id','activity','timestamp','x','y','z']
data4.head()
```

```python
final = pd.concat([data1, data2 , data3 , data4])
```

```python
final.head()
```

```python
null_data = final[final.isnull().any(axis=1)]
null_data
```

```python
final = final.dropna()
```

```python
#removing ';'
df = final.copy()
def removeAfterSC(string):
    """
    input is a string 
    output is a string with everything after comma removed
    """
    return string.split(';')[0].strip()

df['z'] = df['z'].apply(removeAfterSC)

df.info()
```

```python
df['z'] = df['z'].astype(float)
```
